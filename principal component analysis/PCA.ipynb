```python
X = df.drop('goal', axis = 1)
y = df['goal']
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(X)
```

```python
x_train
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 15)
pc = pca.fit(x_train)
```

```python
pc.explained_variance_
```

```python
newpca = PCA(n_components = 10)
new = newpca.fit_transform(x_train)
```

```python
print("orignal shape : ",X.shape)
print("transformed shape : ",new.shape)
```

```python
newdf = pd.DataFrame(data=new,columns=['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10'])
```
